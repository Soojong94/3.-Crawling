### Crawling
- 데이터 수집
- 페이지의 정보를 요청 후 응답 받아 필요한 데이터를 수집하는 행위

#### requests 라이브러리
- 서버에게 페이지정보 (데이터) 를 요청하고 응답 받기 위한 라이브러리
- 브라우저 대신

```python
# 라이브러리 불러오기
import requests as req
```

```python
# 네이버 페이지의 정보를 요청하여 응답받기(requests 라이브러리 활용)
url = 'https://www.naver.com/'
res = req.get(url) # 페이지 정보 가져오기
# response[200] : 성공적으로 페이지 정보를 응답받음
# response[400] : 클라이언트 PC에서 요청상의 문제가 있을 때
# response[500] : 서버pc에서 응답상의 문제가 있을 때
```

```python
# 확인(네이버 페이지의 정보를 잘 받아왔는지)
res.text
# 요청한 페이지의 html 정보를 확인할 수 있음
# 이때 응답받는 데이터의 형태는 문자열
# 데이터를 효율적으로 추출하기 위해서는 html 코드로 변환
```

- 페이지 정보가 업데이트 되면 반드시 다시 페이지 정보를 요청하여 응답받아야 한다

```python
# '푸바오' 를 검색한 페이지의 정보를 요청 응답
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%ED%91%B8%EB%B0%94%EC%98%A4'

res2 = req.get(url)
```

```python
res2.text
# 응답받은 데이터 형태 > 문자열 > html 코드로 변환(태그를 통한 데이터 수집)
```

#### BeautifulSoup 라이브러리⭐⭐⭐
- 컴퓨터가 알아먹을 수 있는 언어로 변경해주는 라이브러리(파싱/parsing)
- 뷰티풀숲 객체화

```python
# 라이브러리 불러오기
from bs4 import BeautifulSoup as bs
```

```python
# bs 객체화
# bs (페이지정보,  파싱방법)
soup = bs (res2.text, 'lxml')
# F12 개발자모드에서의 페이지 정보를 soup이 담고 있음
```

- 네이버 탭 데이터 수집하기

- 요소(element) = 태그 + 콘텐츠


```python
# '이미지' 글자를 추출하기 위해서 태그 선택
soup.select('a')
# 모든 a 태그를 수집
```

```python
# 정확한 태그정보를 선택하는 것이 중요(구분자, 선택자 등을 활용)
# soup.select('태그이름 선택자기호 선택자이름')
data = soup.select('a.tab')
data
# 여러개의 데이터이기 때문에 리스트에 담겨서 출력
```

```python
# 우리가 필요한 순수한 콘텐츠만을 추출 '이미지' << 인덱싱
# 요소.text

data[2].text
```

```python
# 반복문을 통해서 모든 데이터 추출
for i in range(len(data)) :
    print(data[i].text)
```

- 키워드 검색결과 뉴스 제목 수집하기!

```python
# 페이지 정보 'soup' 에서 뉴스 제목 추출하기
# 요소 추출(태그선택)
# 순수한 콘텐츠만을 추출
news = soup.select('a.news_tit')
news[0].text
for i in range(len(news)) :
    print(news[i].text) # 개수가 정해진 경우에 주로 사용하는 반복문
```

```python
for i in news :
    print(i.text) # 개수가 정해지지 않았을 때 사용하는 반복문
```

- 네이버에서 '날씨' 검색후 온도 수집

```python
# 페이지 정보 요청후 응답 받기
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EB%82%A0%EC%94%A8'

res3 = req.get(url)
res3
```

```python
# bs 객체화
soup = bs (res3.text, 'lxml')
soup
```

```python
# 태그 선택
data = soup.select('div.temperature_text>strong')

data
```

```python
# 순수한 콘텐츠 추출
data[0].text
```

- select, select_one 차이점 ⭐⭐⭐
    - select : 해당되는 모든 요소를 추출
    - select_one : 해당되는 가장 상단에 있는 요소 한개만 추출

```python
exdata = soup.select_one('div.temperature_text>strong')
exdata.text
# select_one 기능을 사용하면 요소 형태로 추출되기 때문에
# 리스트 인덱싱 해줄 필요 없이 text 속성 사용 가능!
# 깔끔한 코드 작성을 위해 사용 ("최상단" 데이터 "하나"만 가져와도 될 때)
```
